```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from copy import deepcopy
from bertopic import BERTopic
```

```python
# Load dataset (dataframe)

data = pd.read_csv("Tweets3.csv")
```

```python
# Check the head

data.head()
```

```python
data.tail(20)
```

```python
# Create a list variable containing allthe tweets

docs = list(data.loc[:, "text"].values)
```

```python
# Get the first ten rows of the list

docs[:10]
```

```python
# Model Selection: More than 50 languages are supported

model = BERTopic(language="english")
```

```python
# Fit the model

topics, probs = model.fit_transform(docs)
```

```python
# Extract the most frequent topics

model.get_topic_freq()
```

```python
# Get the individual topics

# first topic:
model.get_topic(0)
```

```python
# Get the second topic
model.get_topic(1)
```

```python
# Get 20th topic
model.get_topic(19)
```

```python
# Visualize the topics

model.visualize_topics()
```
